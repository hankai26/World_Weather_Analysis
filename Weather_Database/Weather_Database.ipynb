```python
import numpy as np
import pandas as pd
from citipy import citipy
import requests
from config import OpenWeather_API_KEY
import json
#import os
#from dotenv import load_dotenv
#OpenWeather_API_KEY = os.getenv("OpenWeather_API_KEY")
```

```python
# Create a set of 2,000 random latitudes and longitudes
lats = np.random.uniform(-90, 90, size=2000)
lngs = np.random.uniform(-180, 180, size=2000)
lat_lng = zip(lats, lngs)
coords = list(lat_lng)
coords[0]
```

```python
# Get the nearest city using the citipy module
cities = []
for coord in coords:
    city = citipy.nearest_city(coord[0], coord[1]).city_name
    if city not in cities:
        cities.append(city)

print(len(cities))
print(cities)
    
```

```python
# Test
#print(json.dumps(city_data, indent=3))
url= "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
url2 = f"{url}appid={OpenWeather_API_KEY}&units={units}&q={cities[0]}"
print(url2)
```

```python
# API call with the OpenWeatherMap
# Retrieve information from API call

url= "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
url2 = f"{url}appid={OpenWeather_API_KEY}&units={units}&q="
city_data = []

for index, city in enumerate(cities):
    print(f"Processing Record{index+1} |{city}")
    query_url = url2+city.replace(" ", "+")
    
    try:
        query_weather = requests.get(query_url).json()
        query_country = query_weather['sys']['country']
        query_lat = query_weather['coord']['lat']
        query_lng = query_weather['coord']['lon']
        query_MaxT = query_weather['main']['temp_max']
        query_hum = query_weather['main']['humidity']
        query_cloud = query_weather['clouds']['all']
        query_wind = query_weather['wind']['speed']
        query_desc = query_weather['weather'][0]['description']
    
        city_data.append({"City": city.title(),
                          "Country": query_country,
                          "Lat": query_lat,
                          "Lng": query_lng,
                          "Max Temp": query_MaxT,
                          "Humidity": query_hum,
                          "Cloudiness": query_cloud,
                          "Wind Speed": query_wind,
                          "Current Description": query_desc})
        print("Found the City. Recording...")
    except:
        print("City not found. Skipping...")
        pass
        
```

```python
# Add data to a new DataFrame
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Save as a CSV file
city_data_df.to_csv("WeatherPy_Database.csv")
```
